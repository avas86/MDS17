### Домашнее задание "Математический анализ. Производная".  
`Уровень 0:`
Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

```python
from scipy.misc import derivative
from sympy import *
import numpy as np
```

```python
x = symbols('x')

def foo(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x+3)

print(derivative(foo, 1.0, dx=1e-4, n=1))
print(derivative(foo, 1.0, dx=1e-4, n=2))
```

`Уровень 1:`
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
import matplotlib.pyplot as plt
```

```python
x_1 = symbols('x')
x_1 = np.linspace(-5, 5, endpoint=True)

def foo_1(x_1):
    return np.sin(2*x_1+1)**5

y1 = foo_1(x_1)
y2 = derivative(foo_1, x_1, dx=1e-4, n=1)
y3 = derivative(foo_1, x_1, dx=1e-4, n=2)

fig, ax = plt.subplots()                        
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.plot(x_1, y1, color="red", label="y(x)")
ax.plot(x_1, y2, color="green", label="y'(x)")
ax.plot(x_1, y3, color="purple", label="y''(x)")
ax.legend()

plt.grid()
plt.show()
```
